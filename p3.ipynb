Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_date(x):
    if x.year > 2000:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.month,x.day)
data['Date'] = data['Date'].apply(fix_date)
data
```

```python
INDEX = ["Min","Max","Mean","Std"]

def min_max_mean_std(row):
    row = pd.Series(row)
    return pd.Series([row.min(), row.max(), row.mean(), row.std()], index=INDEX)


loc_stats = data.apply(min_max_mean_std)
loc_stats
```
